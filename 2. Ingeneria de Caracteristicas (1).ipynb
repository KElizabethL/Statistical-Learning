```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("summer-products-with-rating-and-performance_2020-08.csv")
df.head()
```

```python
categoric_Var=[]
discret_var=[]
Continues_vars= []

for colname in df.columns:
    if(df[colname].dtype == 'object'):
        categoric_Var.append(colname)
    else:
        cantidad_valores = len(df[colname].value_counts())
        if(cantidad_valores<=30):
            discret_var.append(colname)
        else:
            Continues_vars.append(colname)
```

```python
dfcategoric_var = pd.DataFrame(df, columns = categoric_Var)
dfdiscret_var = pd.DataFrame(df, columns = discret_var)
dfcontinuos_var = pd.DataFrame(df, columns = Continues_vars)
```

```python
df_transformado = df[['origin_country','product_variation_size_id','urgency_text','price','retail_price','rating','merchant_rating','units_sold','uses_ad_boosts','badge_product_quality']].copy()
df_transformado
```

```python
def getNanColNames(df):
    colnames = df.columns
    cols_con_na = []
    for col in colnames:
        if(df[col].isnull().sum() > 0):
            cols_con_na.append(col)
    return cols_con_na
cols_con_nan = getNanColNames(df_transformado)
cols_con_nan
df_transformado[cols_con_nan].isnull().mean()
```

```python

df_transformado["product_variation_size_id"].value_counts()
```

```python
df_transformado["origin_country"].value_counts()
```

```python
df_transformado['urgency_text'].fillna('no', inplace=True)
df_transformado['origin_country'].fillna('CN', inplace=True)
df_transformado['product_variation_size_id'].fillna('S', inplace=True)
```

```python
df_transformado[cols_con_nan].isnull().mean()
```

```python
def executeFreqEncoding(df, map_type='freq'):
    colnames = df.columns
    for col in colnames:
        if(df[col].dtype == 'object'):
            factor_div = 1 if (map_type == 'freq') else len(df[col])
            mapper = (df[col].value_counts().sort_values(ascending=False)/factor_div).to_dict()
            df[col] = df[col].map(mapper) 
```

```python
executeFreqEncoding(df_transformado)
df_transformado
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df_transformado)
```
